```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from matplotlib.pyplot import imread, imshow, subplots, show
```

```python
# view image
hw_view = load_img('images/hollywood.jpeg')
hw_view
```

```python
# resize image into new array with 4 dimensions
image = img_to_array(hw_view)

# Creating an array which contains just one image.
image_arr = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
```

```python
# show different shapes
print('Original image size: ', image.shape)
print('Reshaped image size: ', image_arr.shape)
```

```python
def plot_four(image_generator):
    
    image_generator.fit(image_arr)
    img_iterator = image_generator.flow(image_arr)
    
    # plot
    fig, rows = subplots(nrows= 1, ncols= 4, figsize= (20,10))
    for row in rows:
        row.imshow(img_iterator.next()[0].astype('int'))
        row.axis('off')
    show()
```

```python
rotate_gen = ImageDataGenerator(rotation_range= 90)
```

```python
plot_four(rotate_gen)
```

```python
horiz_gen = ImageDataGenerator(horizontal_flip= True)
```

```python
plot_four(horiz_gen)
```

```python
vert_gen = ImageDataGenerator(vertical_flip= True)
```

```python
plot_four(vert_gen)
```

```python
shear_gen = ImageDataGenerator(shear_range= 35)
```

```python
plot_four(shear_gen)
```

```python
width_gen = ImageDataGenerator(width_shift_range= 200)
```

```python
plot_four(width_gen)
```

```python
height_gen = ImageDataGenerator(height_shift_range= 150)
```

```python
plot_four(height_gen)
```

```python
zoom_gen = ImageDataGenerator(zoom_range= [0.1,1.5])
```

```python
plot_four(zoom_gen)
```
